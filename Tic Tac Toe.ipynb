```python
from IPython.display import clear_output

def displayBoard(board):
    clear_output()
    print("Welcome To TIC-TAC-TOE GAME!")
    
    print()
    print()
    print(board[1] + "|" + board[2] + "|" + board[3])
    print("-----")
    print(board[4] + "|" + board[5] + "|" + board[6])
    print("-----")
    print(board[7] + "|" + board[8] + "|" + board[9])
    print()
    print()
    
def markerSelect():
    choice = ""
    while choice not in ["X","O"]:
        
        choice = input("Player1, please choose a marker of your choice!: ")
        if choice not in ["X","O"]:
            print("Invalid MARKER! Please choose again. ")
    print()        
    player1 = choice
    if player1 == "X":
        player2 = "O"
    else:
        player2 = "X"
        
    return player1,player2

def markerPosition():
    
    choice = ""
    while choice not in range(1,10):
        choice  = input("Enter the position between 1-9: ")
        
        if choice.isnumeric():
            if int(choice) not in range(1,10):
                print("Please enter a position within range! ")
            else:
                choice = int(choice)
        else:
            
            print("Please enter valid position")
        
            
            
            
    return choice

    
def checkPosition(board,position):
    if board[position] == " ":
    
        return True
    else:
        print("Element already exist. Try next position")
        return False

def checkWin(board):
    c1 = (board[1] == board[2] == board[3] != " " )
    c2 = (board[4] == board[5] == board[6] != " ")
    c3 =(board[7] == board[8] == board[9] != " ")
    c4= (board[1] == board[4] == board[7] != " ")
    c5 = (board[2] == board[5] == board[8] != " ")
    c6 = (board[3] == board[6] == board[9] != " ")
    c7 = (board[1] == board[5] == board[9] != " ")
    c8 = (board[3] == board[5] == board[7] != " ")
    
    
    return c1 or c2 or c3 or c4 or c5 or c6 or c7 or c8

def playAgain():
    

    choice = input("Do you want a rematch- Yes/No: ")
    return choice == "Yes"
    
    
def game():
    
    play = True
    while play:
    
        board = [" "]*10
        displayBoard(board)
        p1,p2 = markerSelect()

        symbol = p1
        win = False
        for i in range(9):
            if i%2 != 0:
                symbol = p2
                player = "PLAYER2"
            else:
                symbol = p1
                player = "PLAYER1"
            print(f"{player} chance")
            valid = False
            while not valid:
                position = markerPosition()
                valid = checkPosition(board,position)

                if valid:
                    board[position] = symbol
                    break

            displayBoard(board)
            if i >= 4:
                if checkWin(board):
                    win = True
                    print(f"{player} is the winner")
                    break

        if not win:
            print("The game is a draw!")
        
        play = playAgain()
        
    
        
        
game()
```
